## Imprimo % de null y blancos y las variables de cada columna

```python
# Calcular porcentaje de valores nulos por columna
porcentaje_nulos = df.isna().mean() * 100

# Calcular porcentaje de valores en blanco por columna
porcentaje_blancos = df.apply(lambda col: col.astype(str).str.strip().eq('')).mean() * 100

# Calcular número de variables distintas por columna (sin contar NaN ni blancos)
num_variables = df.apply(lambda col: col.dropna().astype(str).str.strip().replace('', pd.NA).dropna().nunique())

# Combinar todo en un solo DataFrame resumen
resumen = pd.DataFrame({
    'Porcentaje_Nulos': porcentaje_nulos.round(2),
    'Porcentaje_Blancos': porcentaje_blancos.round(2),
    'Num_Variables': num_variables
})
print (resumen)
```

```python
# Seleccionar la columna
columna = 'trabajo_de'
serie = df[columna].astype(str).str.strip()

# Eliminar valores en blanco y NaN
serie_limpia = serie.replace('', pd.NA).dropna()

# Contar frecuencia de cada variable
frecuencia = serie_limpia.value_counts()

# Mostrar resultado
print(f"Desglose de la columna '{columna}':")
print(frecuencia)
```
